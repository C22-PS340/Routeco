```python
import os
import pandas as pd
```

```python
polusi17 = ".\SPKU17"
```

```python
files = []
for file in os.listdir(polusi17):
    if file.endswith(".csv"):
        print("Memuat file {0}...".format(file))
        files.append(pd.read_csv(os.path.join(polusi17, file)))

files_gabung = pd.concat(files, axis=0)
files_gabung.to_csv("polusijkt17.csv", index=False)
print("files telah digabungkan")
```

```python
df17 = pd.read_csv("polusijkt17.csv") #index_col = 'tanggal'
df17 = df17.sort_values(["tanggal"], ascending=True)
df17
```

```python
polusi18 = ".\SPKU18"
files = []
for file in os.listdir(polusi18):
    if file.endswith(".csv"):
        print("Memuat file {0}...".format(file))
        files.append(pd.read_csv(os.path.join(polusi18, file)))

files_gabung = pd.concat(files, axis=0)
files_gabung.to_csv("polusijkt18.csv", index=False)
print("files telah digabungkan")
```

```python
dfs = pd.concat([df17, df18])
dfs
```

```python
dfs.info()
```

```python
dfs['stasiun'].unique()
```

```python
dfs.replace('DKI1 (Bunderan HI)','Jakarta Pusat', inplace=True)
dfs.replace('DKI2 (Kelapa Gading)','Jakarta Utara', inplace=True)
dfs.replace('DKI3 (Jagakarsa)', 'Jakarta Selatan', inplace=True)
dfs.replace('DKI4 (Lubang Buaya)','Jakarta Timur', inplace=True)
dfs.replace('DKI5 (Kebon Jeruk)','Jakarta Barat', inplace=True)
dfs.replace('DKI5 (Kebon Jeruk) Jakarta Barat','Jakarta Barat', inplace=True)
dfs['stasiun'].unique()
```
