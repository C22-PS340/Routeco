```python
import os
import pandas as pd
```

```python
data_cuaca = ".\cuaca17-18"
```

```python
files = []
for file in os.listdir(data_cuaca):
    if file.endswith(".csv"):
        print("Memuat file {0}...".format(file))
        files.append(pd.read_csv(os.path.join(data_cuaca, file)))

files_gabung = pd.concat(files, axis=0)
files_gabung.to_csv("cuacajkt17-18.csv", index=False)
print("files telah digabungkan")
```
